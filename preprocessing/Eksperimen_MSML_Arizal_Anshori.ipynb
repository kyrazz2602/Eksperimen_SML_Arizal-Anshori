# **4. Exploratory Data Analysis (EDA)**

Pada tahap ini, Anda akan melakukan **Exploratory Data Analysis (EDA)** untuk memahami karakteristik dataset.

Tujuan dari EDA adalah untuk memperoleh wawasan awal yang mendalam mengenai data dan menentukan langkah selanjutnya dalam analisis atau pemodelan.

```python
target_counts = df['Heart Disease Status'].value_counts()
target_pct    = df['Heart Disease Status'].value_counts(normalize=True) * 100

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Bar chart
axes[0].bar(target_counts.index, target_counts.values,
            color=['#2196F3', '#F44336'], width=0.5,
            edgecolor='white', linewidth=0.8)
axes[0].set_title('Distribusi Heart Disease Status', fontsize=14, fontweight='bold', pad=15)
axes[0].set_xlabel('Heart Disease Status')
axes[0].set_ylabel('Jumlah Pasien')
for i, (label, val) in enumerate(zip(target_counts.index, target_counts.values)):
    axes[0].text(i, val + 50, f'{val:,}', ha='center', fontsize=12, fontweight='bold')

# Pie chart
axes[1].pie(target_counts.values, labels=target_counts.index,
            autopct='%1.1f%%', colors=['#2196F3', '#F44336'],
            startangle=90, wedgeprops={'edgecolor': 'white', 'linewidth': 2})
axes[1].set_title('Proporsi Kelas Target', fontsize=14, fontweight='bold', pad=15)

plt.tight_layout()
plt.show()

print('\n📊 Ringkasan Distribusi Target:')
for label, count, pct in zip(target_counts.index, target_counts.values, target_pct.values):
    print(f'   {label:3s}: {count:,} pasien ({pct:.1f}%)')
print('\n⚠️  Dataset TIDAK seimbang (imbalanced) — kelas Yes hanya 20% dari total data')
```

```python
missing     = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df)) * 100
missing_df  = pd.DataFrame({
    'Jumlah Missing': missing,
    'Persentase (%)': missing_pct.round(2)
}).sort_values('Persentase (%)', ascending=False)
missing_df = missing_df[missing_df['Jumlah Missing'] > 0]

print('📋 Kolom dengan Missing Values:')
print(missing_df.to_string())
```

```python
numeric_cols = ['Age', 'Blood Pressure', 'Cholesterol Level', 'BMI',
                'Sleep Hours', 'Triglyceride Level', 'Fasting Blood Sugar',
                'CRP Level', 'Homocysteine Level']

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    ax = axes[i]
    for status, color in zip(['No', 'Yes'], ['#2196F3', '#F44336']):
        subset = df[df['Heart Disease Status'] == status][col].dropna()
        ax.hist(subset, bins=30, alpha=0.5, color=color,
                label=status, density=True, edgecolor='none')
    ax.set_title(col, fontsize=12, fontweight='bold')
    ax.set_ylabel('Densitas')
    ax.legend(title='Heart Disease', fontsize=8)

plt.suptitle('Distribusi Fitur Numerik berdasarkan Status Penyakit Jantung',
             fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

```python
# Boxplot fitur numerik
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    ax = axes[i]
    data_no  = df[df['Heart Disease Status'] == 'No'][col].dropna()
    data_yes = df[df['Heart Disease Status'] == 'Yes'][col].dropna()
    bp = ax.boxplot([data_no, data_yes], labels=['No', 'Yes'],
                   patch_artist=True, widths=0.5,
                   medianprops=dict(color='black', linewidth=2))
    bp['boxes'][0].set_facecolor('#2196F3')
    bp['boxes'][0].set_alpha(0.6)
    bp['boxes'][1].set_facecolor('#F44336')
    bp['boxes'][1].set_alpha(0.6)
    ax.set_title(col, fontsize=12, fontweight='bold')
    ax.set_xlabel('Heart Disease Status')

plt.suptitle('Boxplot Fitur Numerik berdasarkan Status Penyakit Jantung',
             fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

```python
cat_cols = ['Gender', 'Exercise Habits', 'Smoking', 'Family Heart Disease',
            'Diabetes', 'High Blood Pressure', 'Low HDL Cholesterol',
            'High LDL Cholesterol', 'Alcohol Consumption', 'Stress Level',
            'Sugar Consumption']

fig, axes = plt.subplots(4, 3, figsize=(16, 16))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    ax = axes[i]
    cross = pd.crosstab(df[col], df['Heart Disease Status'],
                        normalize='index') * 100
    cross.plot(kind='bar', ax=ax, color=['#2196F3', '#F44336'],
               edgecolor='white', linewidth=0.5)
    ax.set_title(col, fontsize=12, fontweight='bold')
    ax.set_xlabel('')
    ax.set_ylabel('Proporsi (%)')
    ax.set_xticklabels(ax.get_xticklabels(),
                        rotation=30, ha='right', fontsize=9)
    ax.legend(title='Heart Disease', fontsize=8)

axes[-1].set_visible(False)
plt.suptitle('Proporsi Heart Disease per Kategori Fitur',
             fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

```python
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matriks Korelasi Fitur Numerik', fontsize=14, fontweight='bold', pad=15)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

print('\n🔍 Pasangan fitur dengan korelasi tinggi (|r| > 0.3):')
corr_pairs = corr_matrix.unstack()
corr_pairs = corr_pairs[
    corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1)
]
strong = corr_pairs[abs(corr_pairs) > 0.3].drop_duplicates().sort_values(ascending=False)
if len(strong) > 0:
    print(strong.to_string())
else:
    print('   Tidak ada korelasi kuat antar fitur numerik.')
```

# **5. Data Preprocessing**

Pada tahap ini, data preprocessing adalah langkah penting untuk memastikan kualitas data sebelum digunakan dalam model machine learning.

Jika Anda menggunakan data teks, data mentah sering kali mengandung nilai kosong, duplikasi, atau rentang nilai yang tidak konsisten, yang dapat memengaruhi kinerja model. Oleh karena itu, proses ini bertujuan untuk membersihkan dan mempersiapkan data agar analisis berjalan optimal.

Berikut adalah tahapan-tahapan yang bisa dilakukan, tetapi **tidak terbatas** pada:
1. Menghapus atau Menangani Data Kosong (Missing Values)
2. Menghapus Data Duplikat
3. Normalisasi atau Standarisasi Fitur
4. Deteksi dan Penanganan Outlier
5. Encoding Data Kategorikal
6. Binning (Pengelompokan Data)

Cukup sesuaikan dengan karakteristik data yang kamu gunakan yah. Khususnya ketika kami menggunakan data tidak terstruktur.

```python
# Salin dataset agar data asli tidak berubah
df_clean = df.copy()

n_dup = df_clean.duplicated().sum()
print(f'🔍 Jumlah baris duplikat: {n_dup}')

if n_dup > 0:
    df_clean = df_clean.drop_duplicates()
    print(f'✅ {n_dup} baris duplikat berhasil dihapus.')
    print(f'   Ukuran dataset sekarang: {df_clean.shape}')
else:
    print('✅ Tidak ada data duplikat. Dataset tetap utuh.')
```

```python
# ── Fitur Numerik: Imputasi dengan MEDIAN ──────────────────────────
numeric_impute = ['Age', 'Blood Pressure', 'Cholesterol Level', 'BMI',
                  'Sleep Hours', 'Triglyceride Level', 'Fasting Blood Sugar',
                  'CRP Level', 'Homocysteine Level']

print('📌 Imputasi Fitur Numerik (Median):')
for col in numeric_impute:
    n_missing  = df_clean[col].isnull().sum()
    median_val = df_clean[col].median()
    if n_missing > 0:
        df_clean[col] = df_clean[col].fillna(median_val)
        print(f'   {col:<25s}: {n_missing} missing → diisi median ({median_val:.2f})')

print()
# ── Fitur Kategorikal: Imputasi dengan MODE ─────────────────────────
cat_impute = ['Gender', 'Exercise Habits', 'Smoking', 'Family Heart Disease',
              'Diabetes', 'High Blood Pressure', 'Low HDL Cholesterol',
              'High LDL Cholesterol', 'Alcohol Consumption', 'Stress Level',
              'Sugar Consumption']

print('📌 Imputasi Fitur Kategorikal (Mode):')
for col in cat_impute:
    n_missing = df_clean[col].isnull().sum()
    mode_val  = df_clean[col].mode()[0]
    if n_missing > 0:
        df_clean[col] = df_clean[col].fillna(mode_val)
        print(f"   {col:<25s}: {n_missing} missing → diisi mode ('{mode_val}')")

print()
total_mv = df_clean.isnull().sum().sum()
print(f'✅ Total missing values setelah imputasi: {total_mv}')
```

```python
# ── Ordinal Encoding (Low=0, Medium=1, High=2) ─────────────────────
ordinal_map = {
    'Exercise Habits'    : {'Low': 0, 'Medium': 1, 'High': 2},
    'Alcohol Consumption': {'Low': 0, 'Medium': 1, 'High': 2},
    'Stress Level'       : {'Low': 0, 'Medium': 1, 'High': 2},
    'Sugar Consumption'  : {'Low': 0, 'Medium': 1, 'High': 2},
}

print('📌 Ordinal Encoding (Low=0, Medium=1, High=2):')
for col, mapping in ordinal_map.items():
    df_clean[col] = df_clean[col].map(mapping)
    print(f'   {col:<25s}: {mapping}')

print()

# ── Binary Encoding (No=0, Yes=1) ───────────────────────────────────
binary_cols = ['Smoking', 'Family Heart Disease', 'Diabetes',
               'High Blood Pressure', 'Low HDL Cholesterol', 'High LDL Cholesterol']

print('📌 Binary Encoding (No=0, Yes=1):')
for col in binary_cols:
    df_clean[col] = df_clean[col].map({'No': 0, 'Yes': 1})
    print(f'   {col:<25s}: No=0, Yes=1')

print()

# ── Gender & Target ─────────────────────────────────────────────────
df_clean['Gender'] = df_clean['Gender'].map({'Female': 0, 'Male': 1})
print('📌 Gender Encoding          : Female=0, Male=1')

df_clean['Heart Disease Status'] = df_clean['Heart Disease Status'].map({'No': 0, 'Yes': 1})
print('📌 Target Encoding          : No=0, Yes=1')

print()
print('✅ Encoding selesai. Cuplikan data:')
df_clean.head(3)
```

```python
bins        = [0, 29, 39, 49, 59, 69, 100]
label_codes = [0, 1, 2, 3, 4, 5]
label_names = ['≤29', '30-39', '40-49', '50-59', '60-69', '70+']

df_clean['Age_Group'] = pd.cut(
    df_clean['Age'], bins=bins, labels=label_codes, right=True
).astype(int)

print('📌 Binning Usia:')
print(f'   Batas bin : {bins}')
print(f'   Label     : {dict(zip(label_codes, label_names))}')
print()

age_dist = df_clean['Age_Group'].value_counts().sort_index()
print('   Distribusi Age_Group:')
for code, name, cnt in zip(label_codes, label_names, age_dist.values):
    bar = '█' * (cnt // 100)
    print(f'   {name:>6s} (kode {code}): {cnt:,}  {bar}')

# Visualisasi
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].hist(df['Age'].dropna(), bins=30,
             color='#42A5F5', edgecolor='white', linewidth=0.5)
axes[0].set_title('Distribusi Age (Sebelum Binning)',
                  fontsize=12, fontweight='bold')
axes[0].set_xlabel('Usia (tahun)')
axes[0].set_ylabel('Frekuensi')

colors = ['#1E88E5', '#43A047', '#FB8C00', '#E53935', '#8E24AA', '#00ACC1']
axes[1].bar(label_names, age_dist.values,
            color=colors, edgecolor='white', linewidth=0.5)
axes[1].set_title('Distribusi Age_Group (Setelah Binning)',
                  fontsize=12, fontweight='bold')
axes[1].set_xlabel('Kelompok Usia')
axes[1].set_ylabel('Frekuensi')
for i, v in enumerate(age_dist.values):
    axes[1].text(i, v + 50, f'{v:,}', ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
numeric_scale = ['Age', 'Blood Pressure', 'Cholesterol Level', 'BMI',
                 'Sleep Hours', 'Triglyceride Level', 'Fasting Blood Sugar',
                 'CRP Level', 'Homocysteine Level']

scaler = StandardScaler()

# Simpan nilai asli untuk visualisasi
df_before_scale = df_clean[numeric_scale].copy()

# Terapkan standarisasi
df_clean[numeric_scale] = scaler.fit_transform(df_clean[numeric_scale])

print('✅ Standarisasi Z-Score berhasil diterapkan.')
print('   Formula: z = (x - μ) / σ')
print()
print(f'   {"Kolom":<25s}  {"Mean (sesudah)":>18s}  {"Std (sesudah)":>15s}')
print('   ' + '-' * 62)
for col in numeric_scale:
    m = df_clean[col].mean()
    s = df_clean[col].std()
    print(f'   {col:<25s}  {m:>18.6f}  {s:>15.6f}')
```

```python
# Visualisasi perbandingan sebelum vs sesudah standarisasi
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

for i, col in enumerate(numeric_scale[:6]):
    axes[i].hist(df_before_scale[col], bins=30, alpha=0.6,
                 color='#FF7043', label='Sebelum', density=True, edgecolor='none')
    axes[i].hist(df_clean[col], bins=30, alpha=0.6,
                 color='#42A5F5', label='Sesudah', density=True, edgecolor='none')
    axes[i].set_title(col, fontsize=11, fontweight='bold')
    axes[i].set_ylabel('Densitas')
    axes[i].legend(fontsize=9)

plt.suptitle('Distribusi Fitur Numerik: Sebelum vs Sesudah Standarisasi',
             fontsize=13, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```
